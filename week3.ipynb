```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('telecom_churn.csv')

# Columns based on SID (7th and 8th columns in this case)
column1 = 'Total day calls'
column2 = 'Total day charge'

# Create a bicolour interaction diagram
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x=column1, y=column2, hue='Churn', palette="Set2")
plt.title(f'Bicolour Interaction Diagram: {column1} vs {column2}')
plt.xlabel(column1)
plt.ylabel(column2)
plt.legend(title='Churn')
plt.show()
```
