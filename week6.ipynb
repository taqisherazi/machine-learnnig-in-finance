```python
import pandas as pd
from plotly.offline import iplot
import plotly.graph_objs as go

# Load the dataset
file_path = "XAUUSD_1 Min_Bid_2022.01.01_2022.12.31.csv"
data = pd.read_csv(file_path)

# Extract SID details
SID = 1234567  # Replace with your actual SID
start_point = int(str(SID)[-5:])  # Last 5 digits
time_period = int(str(SID)[-3:])  # Last 3 digits

# Extract the specific portion of the dataset
subset = data.iloc[start_point:start_point + time_period]

# Create the High and Low plot
trace1 = go.Scatter(x=subset.index, y=subset['High'], mode='lines', name='High Price')
trace2 = go.Scatter(x=subset.index, y=subset['Low'], mode='lines', name='Low Price')

layout = go.Layout(title="High and Low Prices",
                   xaxis=dict(title="Time (minutes)"),
                   yaxis=dict(title="Price"),
                   template="plotly")

fig = go.Figure(data=[trace1, trace2], layout=layout)

# Plot the figure
fig.show()

```
